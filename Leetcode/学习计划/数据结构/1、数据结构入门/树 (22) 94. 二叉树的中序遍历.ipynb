# 题目

> 给你二叉树的根节点 `root` ，返回它节点值的**中序遍历**。  
> **中序遍历**是按照访问左子树——根节点——右子树的方式遍历这棵树。

# 方法一：递归

> 中序遍历：左 - 打印 - 右

## 复杂度

- 时间复杂度: $O(n)$ ，其中 $n$ 是二叉树的节点数。

> 每一个节点恰好被遍历一次。

- 空间复杂度: $O(n)$ ，其中 $n$ 是递归中栈帧的开销。

> 平均情况下为 $O(log⁡n)$，最坏情况下树呈现链状，为 $O(n)$ 。

## 代码

### 代码：构建二叉树 

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

### 代码：中序遍历

```python
def inorderTraversal(root):

    res = []
    def dfs(root):
        if not root:
            return
        # 按照 左-打印-右的方式遍历
        dfs(root.left)
        res.append(root.val)
        dfs(root.right)
    dfs(root)
    return res
```

#### 测试一 

```python
root = TreeNode(1)
a = TreeNode(2)
b = TreeNode(3)
c = TreeNode(4)
d = TreeNode(5)
e = TreeNode(6)
```

```python
root.left = a
root.right = b
a.left = c
a.right = d
d.right = e
```

```python
inorderTraversal(root)
```

#### 测试二

```python
root = TreeNode(1)
a = TreeNode(2)
b = TreeNode(3)
```

```python
root.left = a
root.right = b
```

```python
inorderTraversal(root)
```

# 方法二：迭代

> 与递归是等价的，区别在于递归的时候隐式地维护了一个栈，而在迭代的时候需要显式地将这个栈模拟出来，其余的实现与细节都相同。

## 复杂度

- 时间复杂度: $O(n)$ ，其中 $n$ 是二叉树的节点数。

> 每一个节点恰好被遍历一次。

- 空间复杂度: $O(n)$ ，其中 $n$ 是迭代中栈的开销。

> 平均情况下为 $O(log⁡n)$，最坏情况下树呈现链状，为 $O(n)$ 。

## 代码

### 代码：中序遍历

```python
def inorderTraversal(root):

    res = []
    stack = [] #因为中序遍历是后进先出（遍历到最左侧后弹出，然后在向右遍历），因此创建一个栈用于记录
    while stack or root: #只要栈不为空（有节点没被记录），或root不为None（没遍历到叶子节点），则循环继续
        #不断往左子树方向走，每走一次就将当前节点保存到栈中
        #这是模拟递归的调用
        if root: #如果root存在，说明没遍历到叶子节点
            stack.append(root)
            root = root.left
        #当前节点为空，说明左边走到头了，从栈中弹出节点并保存
        #然后转向右边节点，继续上面整个过程
        else: #若root不存在，说明已经遍历到叶子节点，此时需要开始向右遍历
            tmp = stack.pop()
            res.append(tmp.val)
            root = tmp.right #当root为叶子节点时，root = tmp.right为None，此时root不存在。再次循环后，从栈中弹出其父节点，再转向右子节点。
    return res
```

```python
root.left = a
root.right = b
a.left = c
a.right = d
d.right = e
```

# 方法三：Morris 遍历

> 对于每一个当前节点，先找到其左子树的最右侧节点 `pre` 将当前节点及其右子树剪切到 `pre` 上，然后将当前节点设为其左子节点，重复这一过程，直至当前节点没有左子节点，说明已经遍历到最左侧的叶子节点。此时整棵树基本上就变改成了一个链表了，之后再不断往右遍历，将节点加入答案。

## 复杂度

- 时间复杂度: $O(n)$ ，其中 $n$ 是二叉树的节点数。

> 因为 $n$ 个节点的二叉树有 $n−1$ 条边，每条边只可能使用 2 次（一次定位到节点，一次找到前驱节点 `pre` ）。

- 空间复杂度: $O(1)$ 。

> 只操作已经存在的指针（树的空闲指针），因此只需要常数的额外空间。

## 代码

### 代码：中序遍历

```python
def inorderTraversal(root):

    res = []
    pre = None
    while root:
        if root.left: #如果当前节点的左子节点不为空，就将当前节点连带右子树全部挂到左节点的最右子树下面
            pre = root.left
            while pre.right:
                pre = pre.right
            pre.right = root
            #将root指向root的left
            tmp = root
            root = root.left
            tmp.left = None #为了防止出现循环遍历，需要将原先的root的左子树设为空（因为已经在向左遍历了，未来无需再次遍历）
        else: #左子树为空，则打印这个节点，并向右边遍历
            res.append(root.val)
            root = root.right
    return res
```
