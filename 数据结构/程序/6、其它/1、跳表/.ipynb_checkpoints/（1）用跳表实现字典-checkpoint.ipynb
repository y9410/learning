```python
class HeaderNode: #跳表的头节点
    def __init__(self):
        self.next = None
        self.down = None
    
    def getNext(self):
        return self.next
    
    def getDown(self):
        return self.down

    def setNext(self, newnext):
        self.next = newnext

    def setDown(self, newdown):
        self.down = newdown
```

```python
class DataNode:
    def __init__(self, key, value):
        self.key = key
        self.data = value
        self.next = None
        self.down = None
        
    def getKey(self):
        return self.key
    
    def getData(self):
        return self.data
    
    def getNext(self):
        return self.next
    
    def getDown(self):
        return self.down
        
    def setData(self, newdata):
        self.data = newdata
    
    def setNext(self, newnext):
        self.next = newnext
    
    def setDown(self, newdown):
        self.down = newdown
```

```python
class SkipList: #创造一个跳表
    def __init__(self):
        self.head = None
        
    def search(self, key): #搜索某个键
        current = self.head #从头节点开始找
        found = False #是否找到键
        stop = False #是否检查完所有的数据点
        while not found and not stop:
            if current == None:
                stop = True
            else:
                if current.getNext() == None: #如果右侧没有节点，就往下一层找
                    current = current.getDown()
                else:
                    if current.getNext().getKey() == key: #找到目标键
                        found = True
                    else:
                        if key < current.getNext().getKey(): #如果要找的键小于数据节点中的键就说明这一层不会有包含目标键的数据节点
                            current = current.getDown()
                        else: #否则就向右找
                            current = current.getNext()
        if found:
            return current.getNext().getData()
        else:
            return None
        
    def insert(self, key, data):
        if self.head == None: #若没有头节点，则创建一个头节点并将数据节点设置为头节点的next节点
            self.head = HeaderNode()
            temp = DataNode(key, data)
            self.head.setNext(temp)
            top = temp
            while flip() == 1: #只要硬币抛出正面，就新建一层，包括一个头节点和一个数据节点
                newhead = HeaderNode() #新的头节点
                temp = DataNode(key, data) #新的数据节点
                temp.setDown(top) #设置数据节点的down（因为只有一个数据节点因此没有next）
                newhead.setNext(temp) #设置头节点的next和down
                newhead.setDown(self.head)
                self.head = newhead #头节点永远是最上层的新头节点
                top = temp #数据节点永远是最上层的新数据节点
        else: #若已经有了头节点
            towerStack = Stack() #为数据节点的塔创建一个栈
            current = self.head #当前点设置为最上层的头节点
            stop = False
            while not stop:
                if current == None: #若当前点为None，则退出循环
                    stop = True
                else: #否则
                    if current.getNext() == None: #若当前节点的下个节点为None，则把当前节点压入栈（相当于设置一个插入点）
                        towerStack.push(current)
                        current = current.getDown() #转至下一层
                    else: #若当前节点的下个节点不为None
                        if current.getNext().getKey() > key: #若下个数据节点的键大于要插入数据节点
                            towerStack.push(current) #也把当前节点压入栈
                            current = current.getDown() #转至下一层
                        else: #若下个数据节点的键小于要插入数据节点
                            current = current.getNext() #向右继续搜寻
            
            '''循环结束后，在底层插入数据节点'''            
            lowestLevel = towerStack.pop() #栈最上面的点是最底层的插入点
            temp = DataNode(key, data)
            temp.setNext(lowestLevel.getNext()) #将新输入节点的next设置为原插入点的next
            lowestLevel.setNext(temp) #将新数据节点插入插入点的右边
            top = temp
        while flip() == 1: #只要硬币抛出正面
            if towerStack.isEmpty(): #如果栈是空的（说明插入新数据节点前，已有的跳表层数低于新数据节点的塔的层数）
                '''创建新的头节点和数据节点'''
                newhead = HeaderNode()
                temp = DataNode(key, data)
                temp.setDown(top)
                newhead.setNext(temp)
                newhead.setDown(self.head)
                self.head = newhead
                top = temp
            else: #否则（已有的跳表层数高于新数据节点的塔的层数）
                nextLevel = towerStack.pop() #从栈中最上层取出下一个插入点
                temp = DataNode(key, data)
                temp.setDown(top) #将新数据节点插入到取出的插入点中
                temp.setNext(nextLevel.getNext())
                nextLevel.setNext(temp)
                top = temp
```

```python
from random import randrange
def flip():
    return randrange(2)
```

```python
class Map: #用跳表实现映射类
    def __init__(self):
        self.collection = SkipList()
    
    def put(self, key, value):
        self.collection.insert(key, value)
    
    def get(self, key):
        return self.collection.search(key)
```

```python
#假设为n个键构建跳表，有一半键高度为2，有四分之一键高度为4...因此跳表的最大高度是(log 2 n)+ 1，高度记为为O(logn)
#最坏情况下需要查找O(logn)层，在每一层中有一半概率向右查找，一半概率向下查找，因此每一层查找次数是常数，因此查找和插入（主要是查找插入位置）的复杂度为O(logn)
```
