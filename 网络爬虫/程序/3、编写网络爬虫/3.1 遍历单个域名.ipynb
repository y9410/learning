```python
#之所以叫网络爬虫，是因为它们可以在 Web 上爬行。它们本质上就是一种递归方式
#它们必须首先获取一个 URL 对应的网页内容，然后检查这个页面，寻找另一个 URL，再获取该 URL 对应的网页内容，并不断循环这一过程
```

```python
#使用网络爬虫的时候，必须非常谨慎地考虑需要消耗多少带宽，还要尽力思考能不能让抓取目标的服务器负载更低一些
```

```python
import urllib
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re
```

```python
#设置代理服务器
proxy = "http://10.204.32.132:7890" #代理服务器名：端口
proxy_py = urllib.request.ProxyHandler({'https':proxy})  #设置python认识的代理：默认为null
opener = urllib.request.build_opener(proxy_py)           #建立
urllib.request.install_opener(opener)                  #运行代理
```

```python
#获取维基百科Kevin_Bacon页面中的所有链接
html = urlopen('https://en.wikipedia.org/wiki/Kevin_Bacon') #要爬取的网页（一定要使用"https"）
bs = BeautifulSoup(html, 'html.parser')
```

```python
for link in bs.find_all('a'):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
#直接爬取页面会获得一些不需要的链接
#只需要包含词条的页面的链接
#指向词条页面（不是指向其他内部页面）的链接存在三个共同点：它们都在 id 是 bodyContent 的 div 标签里、URL 不包含冒号、URL 都以 /wiki/ 开头
```

```python
for link in bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$')):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
#将上述功能模块化
```

```python
random.seed(datetime.datetime.now())
def getLinks(articleUrl):
    html = urlopen('https://en.wikipedia.org{}'.format(articleUrl)) #要爬取的网页（一定要使用"https"）
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a',href=re.compile('^(/wiki/)((?!:).)*$'))
```

```python
links = getLinks('/wiki/Kevin_Bacon')
```

```python
while len(links) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```
